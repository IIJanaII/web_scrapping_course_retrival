```python
def map_price(price):
    if price == 'No price':
        return np.nan
    elif 'Free' in price:
        return 0
    elif 'Paid' in price:
        return np.nan
    else:
        # Extract numeric values from the string and convert to integer
        return int(''.join(filter(str.isdigit, price)))

# Apply the mapping function to create the new column
df['new_price_column'] = df['price'].map(map_price)
```
